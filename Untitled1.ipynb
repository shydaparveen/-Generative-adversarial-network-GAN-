```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
print(list(zip(x,y)))

mean_y=sum(y)/len(y)
mean_x=sum(x)/len(x)
numerator = sum([(xi - mean_x) * (yi - mean_y) for xi,yi in zip(x,y)])
denominator_x = sum([(xi - mean_x) ** 2 for xi in x])
denominator_y = sum([(yi - mean_y) ** 2 for yi in y])
correlation_coefficient= numerator /((denominator_x * denominator_y)**0.5)
print("correlation_coefficient betwwen 'x' and 'y':",correlation_coefficient)
print(numerator)
print(denominator_x)
print(denominator_y)
```
