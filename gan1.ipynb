```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
num_students = 50
exam1_scores = np.random.normal(70,10,num_students)
exam2_scores = np.random.normal(75,15,num_students)
```

```python
exam1_scores
```

```python
plt.hist(exam1_scores)
```

```python
plt.hist(exam2_scores)
```

```python
exam1_scores.mean()
```

```python
exam2_scores.mean()
```

```python
plt.boxplot(exam1_scores)
```

```python
exam1_scores[0] = 120
exam2_scores[1] = 3
plt.figure(figsize=(8,6))
plt.scatter(exam1_scores, exam2_scores)
plt.title('Scatter Plot of Exam Scores')
plt.xlabel('Éxam 1 Score')
plt.xlabel('Éxam 2 Score')
plt.grid(True)
plt.show()


```
